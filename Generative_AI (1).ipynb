Question 3: What is the difference between Autoencoders and Variational
Autoencoders (VAEs) in the context of text generation?
  - Autoencoders:
    - An Autoencoder is a neural network that learns to compress input data into a smaller representation (latent vector) and then reconstruct the original input from it.
    - Working:
    - Encoder: Compresses the input text into a latent vector.
Latent Space: Stores the compressed representation.
    - Decoder: Reconstructs the original text from the latent vector.
    - Characteristics:
      - Learns a fixed latent representation.
      - Mainly used for data compression, feature extraction, and noise removal.
      - Not suitable for generating new and diverse text because the latent space is not structured for sampling.
  - Variatinal Autoencoders (VAEs)
     - A Variational Autoencoder (VAE) is an improved version of the Autoencoder. Instead of learning a single latent vector, it learns a probability distribution (mean and variance) for the latent space.
     - Working:
     - Encoder: Produces the mean (μ) and variance (σ²) of the latent distribution.
    - Sampling: A latent vector is randomly sampled from this distribution.
    - Decoder: Generates or reconstructs text from the sampled latent vector.
   - Characteristics:
     - Learns a probabilistic latent space.
     - Can generate new and diverse text samples.
     - Produces smoother and more meaningful latent representations.
     - Suitable for text generation, sentence generation, and data augmentation.

Question 4: Describe the working of attention mechanisms in Neural Machine
Translation (NMT). Why are they critical?
  - Attention Mechanism in Neural Machine Translation (NMT)
    - The Attention Mechanism is a technique used in Neural Machine Translation (NMT) that allows the model to focus on the most relevant words in the input sentence while translating each word in the output sentence.
    - In traditional Encoder–Decoder models, the encoder compresses the entire input sentence into a single fixed-length vector. For long sentences, this vector may lose important information, leading to poor translations.
    - The attention mechanism solves this problem by allowing the decoder to look back at all encoder hidden states and assign different importance (attention weights) to each input word during translation.

Answer 6.1: 1. Explanation of the ApproachA Variational Autoencoder differs from a standard Autoencoder by mapping inputs to a probabilistic distribution rather than a fixed vector.
  - Encoder:
    - Tokenizes and embeds the text, then compresses it into two learned parameters: a mean ($\mu$) and a log-variance ($\log\sigma^2$).
  - Reparameterization Trick:
    - We sample a latent vector $z$ using $z = \mu + \epsilon \cdot \sigma$, where $\epsilon \sim \mathcal{N}(0, 1)$. This allows gradients to backpropagate through the random sampling process.
  - Decoder:
    - Takes the sampled latent vector $z$ and reconstructs the original sentence by predicting the probability distribution of words at each sequence position.
  - Loss Function:
    - The VAE is optimized using a combined loss function:$$\mathcal{L} = \text{Reconstruction Loss} + D_{KL}(q_\phi(z|x) || p_\theta(z))$$(Cross-Entropy Loss for accurate word prediction + KL Divergence to regularize the latent space to a standard normal distribution).

```python
# Answer 6.2
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import numpy as np

# ==========================================
# 1. Preprocess the Data
# ==========================================
dataset = [
    "The sky is blue",
    "The sun is bright",
    "The grass is green",
    "The night is dark",
    "The stars are shining"
]

# Lowercase and split into words
tokenized_data = [sentence.lower().split() for sentence in dataset]

# Build vocabulary
vocab = set(word for sentence in tokenized_data for word in sentence)
word2idx = {word: idx + 1 for idx, word in enumerate(vocab)}
word2idx["<PAD>"] = 0  # Padding token
idx2word = {idx: word for word, idx in word2idx.items()}
vocab_size = len(word2idx)

# Pad sequences to max length (5 words)
max_seq_len = max(len(seq) for seq in tokenized_data)

def encode_sequence(seq):
    padded = seq + ["<PAD>"] * (max_seq_len - len(seq))
    return [word2idx[w] for w in padded]

# Convert to PyTorch Tensors
tensor_data = torch.tensor([encode_sequence(seq) for seq in tokenized_data], dtype=torch.long)


# ==========================================
# 2. Build the Basic VAE Model
# ==========================================
class SimpleTextVAE(nn.Module):
    def __init__(self, vocab_size, seq_len, embed_dim=16, hidden_dim=32, latent_dim=8):
        super(SimpleTextVAE, self).__init__()
        self.seq_len = seq_len
        self.vocab_size = vocab_size

        # Embedding Layer
        self.embedding = nn.Embedding(vocab_size, embed_dim, padding_idx=0)

        # Encoder
        self.enc_fc1 = nn.Linear(seq_len * embed_dim, hidden_dim)
        self.enc_mu = nn.Linear(hidden_dim, latent_dim)
        self.enc_logvar = nn.Linear(hidden_dim, latent_dim)

        # Decoder
        self.dec_fc1 = nn.Linear(latent_dim, hidden_dim)
        self.dec_fc2 = nn.Linear(hidden_dim, seq_len * vocab_size)

    def encode(self, x):
        # Embed and flatten the sequence
        embedded = self.embedding(x).view(x.size(0), -1)
        hidden = torch.relu(self.enc_fc1(embedded))
        return self.enc_mu(hidden), self.enc_logvar(hidden)

    def reparameterize(self, mu, logvar):
        std = torch.exp(0.5 * logvar)
        eps = torch.randn_like(std)
        return mu + eps * std

    def decode(self, z):
        hidden = torch.relu(self.dec_fc1(z))
        output = self.dec_fc2(hidden)
        # Reshape to (Batch, Seq_Len, Vocab_Size)
        return output.view(-1, self.seq_len, self.vocab_size)

    def forward(self, x):
        mu, logvar = self.encode(x)
        z = self.reparameterize(mu, logvar)
        reconstruction = self.decode(z)
        return reconstruction, mu, logvar

def loss_function(recon_x, x, mu, logvar):
    # Flatten the outputs for CrossEntropyLoss
    recon_loss = F.cross_entropy(
        recon_x.view(-1, vocab_size),
        x.view(-1),
        reduction='sum'
    )
    # KL Divergence
    kl_divergence = -0.5 * torch.sum(1 + logvar - mu.pow(2) - logvar.exp())
    return recon_loss + kl_divergence


# ==========================================
# 3. Train the Model
# ==========================================
model = SimpleTextVAE(vocab_size, max_seq_len)
optimizer = optim.Adam(model.parameters(), lr=0.01)
epochs = 300

print("Training VAE...")
for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()

    recon_batch, mu, logvar = model(tensor_data)
    loss = loss_function(recon_batch, tensor_data, mu, logvar)

    loss.backward()
    optimizer.step()

    if (epoch + 1) % 100 == 0:
        print(f"Epoch {epoch+1}/{epochs} | Loss: {loss.item():.4f}")


# ==========================================
# 4. Reconstruction and Generation
# ==========================================
print("\n--- Model Outputs ---")

def decode_output(tensor_output):
    # Get the index of the highest probability word
    predicted_indices = torch.argmax(tensor_output, dim=-1)
    sentences = []
    for seq in predicted_indices:
        words = [idx2word[idx.item()] for idx in seq if idx.item() != 0] # Ignore <PAD>
        sentences.append(" ".join(words))
    return sentences

model.eval()
with torch.no_grad():
    # 1. Reconstruction Test
    recon_output, _, _ = model(tensor_data)
    reconstructed_texts = decode_output(recon_output)

    print("\n[Reconstruction]")
    for i, orig in enumerate(dataset):
        print(f"Original: {orig.lower()}")
        print(f"Reconstructed: {reconstructed_texts[i]}\n")

    # 2. Generation Test (Sampling from latent space)
    print("[Generation from Random Latent Space]")
    random_z = torch.randn(3, 8) # 3 samples, latent_dim = 8
    generated_output = model.decode(random_z)
    generated_texts = decode_output(generated_output)

    for i, text in enumerate(generated_texts):
        print(f"Generated {i+1}: {text}")
```

```python
import os
from openai import OpenAI

# Initialize the OpenAI client
# Note: You need to set your API key as an environment variable:
# os.environ["OPENAI_API_KEY"] = "your-api-key-here"
client = OpenAI(api_key=os.environ.get("OPENAI_API_KEY"))

# 1. Define the original text
original_text = (
    "The advancement of artificial intelligence has transformed the way we interact "
    "with technology. It offers new solutions to complex problems, making our daily "
    "tasks more efficient and opening doors to unprecedented innovations."
)

# 2. Define a translation function
def translate_text(text, target_language):
    """Translates text to the specified target language using GPT-3.5."""
    prompt = f"Translate the following English text to {target_language}:\n\n{text}"

    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "You are a highly skilled translation assistant."},
            {"role": "user", "content": prompt}
        ],
        temperature=0.3, # Low temperature for accurate, deterministic translation
        max_tokens=200
    )

    return response.choices[0].message.content.strip()

# 3. Execute translations
print("Translating...\n")
french_translation = translate_text(original_text, "French")
german_translation = translate_text(original_text, "German")

# 4. Output the results
print("--- Original English Text ---")
print(original_text)
print("\n--- French Translation ---")
print(french_translation)
print("\n--- German Translation ---")
print(german_translation)
```

Question 8: Implement a simple attention-based encoder-decoder model for
English-to-Spanish translation using Tensorflow or PyTorch.
  - Architecture Overview
    - Encoder:
      - Reads the source sentence (English) word by word and processes it using a Recurrent Neural Network (like a GRU). It outputs a sequence of hidden states.
    - Attention Mechanism:
      - Instead of forcing the encoder to compress the whole sentence into a single vector, attention allows the decoder to "look back" at specific parts of the encoder's outputs (the source sentence) at each step of the translation.
    - Decoder:
      - Generates the target sentence (Spanish) one word at a time. At each step, it uses its current state, the previously generated word, and the context vector provided by the attention mechanism to predict the next word.

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import numpy as np

# ==========================================
# 1. Toy Dataset and Preprocessing
# ==========================================
# Small English-to-Spanish pairs
data = [
    ("i love you", "te amo"),
    ("he is good", "el es bueno"),
    ("hello world", "hola mundo"),
    ("good morning", "buenos dias")
]

# Vocabulary builder
def build_vocab(sentences):
    vocab = {"<PAD>": 0, "<SOS>": 1, "<EOS>": 2} # SOS: Start of Sequence, EOS: End
    for sentence in sentences:
        for word in sentence.split():
            if word not in vocab:
                vocab[word] = len(vocab)
    return vocab

eng_sentences = [pair[0] for pair in data]
spa_sentences = [pair[1] for pair in data]

eng_vocab = build_vocab(eng_sentences)
spa_vocab = build_vocab(spa_sentences)
spa_idx2word = {idx: word for word, idx in spa_vocab.items()}

# Convert text to tensor sequences
def text_to_tensor(sentence, vocab, max_len=5):
    indices = [vocab.get(word, 0) for word in sentence.split()]
    indices.append(vocab["<EOS>"]) # Append End-of-Sequence
    # Pad to max_len
    padded = indices + [vocab["<PAD>"]] * (max_len - len(indices))
    return torch.tensor(padded[:max_len], dtype=torch.long)

# Batch size of 4
X = torch.stack([text_to_tensor(eng, eng_vocab) for eng in eng_sentences])
Y = torch.stack([text_to_tensor(spa, spa_vocab) for spa in spa_sentences])


# ==========================================
# 2. Model Definitions
# ==========================================

class Encoder(nn.Module):
    def __init__(self, input_size, hidden_size):
        super().__init__()
        self.embedding = nn.Embedding(input_size, hidden_size)
        self.gru = nn.GRU(hidden_size, hidden_size, batch_first=True)

    def forward(self, x):
        embedded = self.embedding(x)
        outputs, hidden = self.gru(embedded)
        # outputs: (batch, seq_len, hidden_size)
        # hidden: (1, batch, hidden_size)
        return outputs, hidden


class BahdanauAttention(nn.Module):
    def __init__(self, hidden_size):
        super().__init__()
        self.Wa = nn.Linear(hidden_size, hidden_size)
        self.Ua = nn.Linear(hidden_size, hidden_size)
        self.Va = nn.Linear(hidden_size, 1)

    def forward(self, query, keys):
        # query (decoder hidden state): (batch, 1, hidden_size)
        # keys (encoder outputs): (batch, seq_len, hidden_size)

        scores = self.Va(torch.tanh(self.Wa(query) + self.Ua(keys))) # (batch, seq_len, 1)
        weights = F.softmax(scores, dim=1) # Attention weights

        # Multiply weights by encoder outputs to get context vector
        context = torch.bmm(weights.transpose(1, 2), keys) # (batch, 1, hidden_size)
        return context, weights


class AttentionDecoder(nn.Module):
    def __init__(self, hidden_size, output_size):
        super().__init__()
        self.embedding = nn.Embedding(output_size, hidden_size)
        self.attention = BahdanauAttention(hidden_size)
        self.gru = nn.GRU(hidden_size * 2, hidden_size, batch_first=True)
        self.out = nn.Linear(hidden_size, output_size)

    def forward(self, x, hidden, encoder_outputs):
        # x is the previous word index: (batch, 1)
        embedded = self.embedding(x)

        # Query for attention is the previous hidden state
        query = hidden.transpose(0, 1)
        context, attn_weights = self.attention(query, encoder_outputs)

        # Combine embedded input word and context vector
        rnn_input = torch.cat((embedded, context), dim=2)

        output, hidden = self.gru(rnn_input, hidden)
        prediction = self.out(output.squeeze(1)) # (batch, output_vocab_size)

        return prediction, hidden


# ==========================================
# 3. Initialization and Training
# ==========================================
HIDDEN_SIZE = 32
LEARNING_RATE = 0.01
EPOCHS = 200

encoder = Encoder(len(eng_vocab), HIDDEN_SIZE)
decoder = AttentionDecoder(HIDDEN_SIZE, len(spa_vocab))

criterion = nn.CrossEntropyLoss(ignore_index=spa_vocab["<PAD>"])
encoder_optimizer = optim.Adam(encoder.parameters(), lr=LEARNING_RATE)
decoder_optimizer = optim.Adam(decoder.parameters(), lr=LEARNING_RATE)

print("Training Attention Seq2Seq...")
for epoch in range(EPOCHS):
    encoder_optimizer.zero_grad()
    decoder_optimizer.zero_grad()

    loss = 0
    batch_size = X.size(0)
    target_len = Y.size(1)

    # 1. Pass input through Encoder
    encoder_outputs, hidden = encoder(X)

    # 2. Prepare Decoder input (<SOS> token)
    decoder_input = torch.tensor([[spa_vocab["<SOS>"]]] * batch_size)

    # 3. Decode step by step (Teacher Forcing)
    for t in range(target_len):
        prediction, hidden = decoder(decoder_input, hidden, encoder_outputs)

        # Calculate loss for this time step
        loss += criterion(prediction, Y[:, t])

        # Next input is the true target word (Teacher Forcing)
        decoder_input = Y[:, t].unsqueeze(1)

    loss.backward()
    encoder_optimizer.step()
    decoder_optimizer.step()

    if (epoch + 1) % 50 == 0:
        print(f"Epoch {epoch+1}/{EPOCHS} | Loss: {loss.item() / target_len:.4f}")

# ==========================================
# 4. Evaluation / Inference
# ==========================================
print("\n--- Translation Results ---")
encoder.eval()
decoder.eval()

with torch.no_grad():
    encoder_outputs, hidden = encoder(X)
    decoder_input = torch.tensor([[spa_vocab["<SOS>"]]] * X.size(0))

    decoded_words = [[] for _ in range(X.size(0))]

    # Generate words up to max length
    for t in range(5):
        prediction, hidden = decoder(decoder_input, hidden, encoder_outputs)

        # Get highest probability word
        top_idx = prediction.argmax(dim=1)

        for i, idx in enumerate(top_idx.tolist()):
            decoded_words[i].append(spa_idx2word[idx])

        # Next input is the network's own prediction
        decoder_input = top_idx.unsqueeze(1)

    # Clean up output formatting
    for i, seq in enumerate(decoded_words):
        # Stop at <EOS>
        clean_seq = []
        for word in seq:
            if word == "<EOS>": break
            if word not in ["<PAD>", "<SOS>"]: clean_seq.append(word)

        print(f"English: {eng_sentences[i]}")
        print(f"Spanish: {' '.join(clean_seq)}\n")
```

Explanation of the Approach:

```python
from transformers import pipeline, set_seed

# ==========================================
# 1. Initialize the Model
# ==========================================
# Load the pre-trained GPT-2 model for text generation
print("Loading GPT-2 model...")
generator = pipeline('text-generation', model='gpt2')

# Set a seed for reproducibility (optional)
set_seed(42)

# ==========================================
# 2. Construct the Prompt (Simulated Fine-Tuning)
# ==========================================
# We use the dataset as context to show the model the desired pattern
dataset_context = (
    "Roses are red, violets are blue,\n"
    "Sugar is sweet, and so are you.\n\n"
    "The moon glows bright in silent skies,\n"
    "A bird sings where the soft wind sighs.\n\n"
)

# Instructions to trigger the generation of a new poem
prompt = dataset_context + "Here is a new short poem in the same style:\n"

# ==========================================
# 3. Generate the Poem
# ==========================================
print("Generating poem...\n")
output = generator(
    prompt,
    max_new_tokens=25,       # Enough for 2-4 short lines
    temperature=0.7,         # Balance between creativity and structure
    num_return_sequences=1,
    pad_token_id=50256       # Suppresses the open-end generation warning
)

# ==========================================
# 4. Clean and Display the Output
# ==========================================
generated_text = output[0]['generated_text']

# Extract only the newly generated part (removing the prompt)
new_poem_raw = generated_text[len(prompt):].strip()

# Split by lines and take the first 2 to 4 lines to ensure it fits the constraints
lines = new_poem_raw.split('\n')
clean_poem = "\n".join([line for line in lines if line.strip()][:4])

print("--- The Prompt Used ---")
print(prompt)

print("--- Generated Poem ---")
print(clean_poem)
```

3. Real-World Challenges
  - The "Cliché Convergence":
    - Left unguided, LLMs optimize for statistical averages, generating generic, predictable plots (e.g., a mysterious orphan discovering magic powers). Maintaining high novelty requires carefully tuning generation randomness and constraints.
  - Long-Term Context Drift:
    - Novels or long outlines span thousands of words. Models often suffer from memory decay, changing a character's physical description, backstory, or internal logic midway through a plot.
  - Copyright and Plagiarism Risks:
    - Generative models can accidentally memorize unique phrasing or distinctive intellectual properties from their training data, exposing the publishing house to legal liabilities.
